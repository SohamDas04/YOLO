```python
import pandas as pd
import numpy as np
performance2=pd.read_csv('results copy.csv')
performance2
```

```python
final_performance=pd.concat([performance1,performance2],ignore_index=True)
final_performance
```

```python
box_loss=final_performance['         train/box_loss'].to_numpy()
class_loss=final_performance['         train/cls_loss'].to_numpy()
epochs = range(1,30+1)
box_loss
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(epochs, box_loss, 'b', label='Box Loss')
plt.title('Box Loss vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Box Loss')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(epochs, class_loss, 'b', label='Class Loss')
plt.title('Class Loss vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Class Loss')
plt.show()
```
